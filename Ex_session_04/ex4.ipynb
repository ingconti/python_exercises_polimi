## ex 4.1

Prepare on you PC in the SAME directory of our python program
2 or more files composed by multiple integer values, comma separated
like that: 

120,230,100,400,500  called 
samples_001.csv, samples_002.csv and so on,
where 001 002 are the patient code.

Write a program that receives in input a string representing the patent code, check if file exists, and prints the average of values,
or False if file cannot be open or contents is invalid.

```python
FILE_PREFIX = "samples_"
FILE_SUFFIX = ".csv"


def buildName(patientCode):
	fname = FILE_PREFIX + patientCode + FILE_SUFFIX
	return fname


def avgFrom(line):
	values = line.split(',')
	valuesCount = len(values)
	if valuesCount > 0:
		sum = int(values[0])
		for i in range(1, valuesCount):
			value = int(values[i])
			sum += value
		return float(sum) / valuesCount  # pass to float with right conversion
	else:
		return 0.0

def process(fileName):
	try:
		f = open(fileName)
	except:
		return False
	else:
		lines = f.read().splitlines()
		f.close()
		lineCount = len(lines)
		if lineCount > 0:
			line = lines[0]  # take first only
			avg = avgFrom(line)
			print(avg)
			return True
			
		return False
		

patientCode = input('Enter patient''s code: ')
fileName = buildName(patientCode)
result = process(fileName)
if process(fileName) == False:
	print(False)
```

## Ex 4.2

In Your folder there are many text files containing data of surgeries. Those files are added and deleted when needed.
Files are named with bed number ranging from 01 to 30.
Write a program that reads existing files, prints file names and contents.

(Program must be written using subprograms extensively.)


Every file has the following structure:
  name: John
  surname: Doe
  surgery_date:2020-11-21

<b>Note:</b>
    in some files you can have:

surname: Smith
name: Mark
surgery_date:2020-11-23

but program must:
<ul>
<li>write fields in the correct order</li>
<li>replace numeric month with its English name.</li>
</ul>


Expected output:

file name:  01.txt

Doe John 2020-July-28 

file name:  05.txt

Smith Mark 2020-November-24


(run note: You wuold see data for bed 1 and bed 5)

```python
MAX_BEDS = 10
MONTHS = ["January", "February", "March","April",
		  "May", "June", "July", "August",
		  "September", "October", "November", "December"]
MONTHS_NUM = 12

FILE_PREFIX = "bed_"
FILE_SUFFIX = ".txt"

def buildName(bedNumber):
	bedNumberStr = str(bedNumber)
	if len(bedNumberStr) < 2:
		bedNumberStr = FILE_PREFIX + '0' + bedNumberStr
	fname = bedNumberStr + FILE_SUFFIX
	return fname


#if not found, return ""
def findDataFor(fieldName, inLines):
	result = ""
	l = len(inLines)
	i = 0
	found = False

	while i < l and not found:
		info = inLines[i].split(':')
		if info[0] == fieldName:
			found = True
		else:
			i += 1
	if found:
		result = info[1].strip()
	return  result


def convertDate(s):
	result = "invalid month"
	#get month as index:
	yearStr = s[0:4]
	monthStr = s[5:7]
	dayStr = s[8:10]
	if monthStr.isdigit():
		index = int(monthStr)  - 1
		if index < MONTHS_NUM:
			monthName = MONTHS[index]

	result = yearStr + '-' + monthName + '-' + dayStr
	return result

def printPatientSurgeryData(lines):
	#use  array of field names to parse rows.
	keys = ['surname',  'name', 'surgery_date']
	patientRecord = ""
	for k in keys:
		s = findDataFor(k, lines)
		if len(s)>0: # test if we cannot find a valid field.
			if k != "surgery_date" :
				patientRecord = patientRecord + s + " "
			else:
				patientRecord = patientRecord + convertDate(s) + " "

	print(patientRecord)

def process(fileName):
	# print('processing ', fileName)
	try:
		f = open(fileName)		
	except:
		return False
	else:
		print('file name: ', fileName)
		lines = f.read().splitlines()
		printPatientSurgeryData(lines)
		f.close()
		
for bedNumber in range(1,MAX_BEDS):
	fileName = buildName(bedNumber)
	result = process(fileName)

```

<b>Write a program that open all files and prints the width (end-start)

```python
import pandas as pd

INFO_FILE_NAME = "info.txt"
N_OF_SAMPLES_KEY = "Number of samples"

FILE_PREFIX = "S_"
FILE_SUFFIX = ".gdm"

THRESHOLD = 0.19

def readInfoFile():
	try:
		dataFrame = pd.read_csv(INFO_FILE_NAME, sep = '\t', header = None)
		result = int(dataFrame.loc[5,1])
	except:
		result = 0
	return result

def process(nofFiles):
	for i in range(0,nOfFiles):
		fname = buildFileName(i)
		processFile(fname)

def buildFileName(index):
	s = FILE_PREFIX + '{:0>5}'.format(index)+FILE_SUFFIX
	return s


def processFile(fname):
	print(fname)
	dataFrame = pd.read_csv(fname, sep = '\t',  header = None)

	dataFrame.columns = ['chr', 'start', 'end', 'strand', 'source', 'feature',
						 'score', 'frame', 'name', 'signal',
						'pvalue', 'qvalue', 'peak'
						 ]

	dataFrame['pvalue'] = pd.to_numeric(dataFrame['pvalue'])

	filtered = dataFrame[dataFrame['pvalue']/100 > THRESHOLD]

	width_df = filtered['end'] -  filtered['start']
	for v in width_df:
		print(v)

nOfFiles = readInfoFile()
if nOfFiles>0:
	print(nOfFiles ," to be processed")
	process(nOfFiles)
else:
	print(INFO_FILE_NAME + " is missing")
```

## Ex4.4 GMQL data manipulation with files functions

Rewrite the above program using standard file functions making extensive use of subprograms.




```python
INFO_FILE_NAME = "info.txt"
N_OF_SAMPLES_KEY = "Number of samples"

FILE_PREFIX = "S_"
FILE_SUFFIX = ".gdm"
THRESHOLD = 0.19

def readFileLines(fileName):
	try:
		f = open(fileName)
	except:
		return []
	else:
		lines = f.read().splitlines()
		return lines


def readInfoFile():
	result = 0
	lines = readFileLines(INFO_FILE_NAME)
	#scan for N_OF_SAMPLES_KEY:
	for l in lines:
		(key, value) = l.split('\t')
		if key ==  N_OF_SAMPLES_KEY:
			result = int(value)
			break

	return result

def buildFileName(index):
	s = FILE_PREFIX + '{:0>5}'.format(index)+FILE_SUFFIX
	return s


def processLine(l):
	#print(l)
	parts = l.split('\t')
	#print(len(parts))
	end = int(parts[2])
	start = int(parts[1])
	pValue = float(parts[10])
	width = end - start
	if pValue /100 > THRESHOLD:
		print(width)

def processFile(fname):
	print(fname)
	lines = readFileLines(fname)
	lineCount = len(lines)
	if lineCount > 0:
		for l in lines:
			processLine(l)
	else:
		print("cannot open "  + fname)

def process(nofFiles):
	for i in range(0,nOfFiles):
		fname = buildFileName(i)
		processFile(fname)
#main:
nOfFiles = readInfoFile()
if nOfFiles>0:
	print(nOfFiles ," to be processed")
	process(nOfFiles)
else:
	print(INFO_FILE_NAME + " is missing")
```
